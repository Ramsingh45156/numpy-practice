#create Numpy array using numpy function

```python
import numpy as np
```

```python
#zero 0's array

arr=np.zeros(4)
print(arr)
print(type(arr))
print(arr.dtype)
print("your array Dimension:-",arr.ndim)
```

```python
arr=np.zeros((4,3))
print(arr)
print(type(arr))
print(arr.dtype)
print("your array Dimension:-",arr.ndim)
```

```python
#(4,3,2) matlab:
#4 blocks
#har block me 3 rows
#har row me 2 elements

arr=np.zeros(((4,3,2)))
print(arr)
print(type(arr))
print(arr.dtype)
print("your array Dimension:-",arr.ndim)
```
