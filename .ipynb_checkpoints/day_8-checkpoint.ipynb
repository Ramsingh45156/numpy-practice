#function change

```python
x2=np.array([1,2,3,4,5])
new=np.float32(x2)
print(x2)
print("Data type:",x2.dtype)
print(new)
print("New Data type:",new.dtype)
```

```python
x2=np.array([1,2,3,4,5])
new=np.float32(x2)
new_one=np.int_(new)
print(x2)
print("Data type:",x2.dtype)
print(new)
print("New Data type:",new.dtype)
print(new_one)
print("New_one Data type:",new_one.dtype)
```

```python
x3=np.array([1,2,3,4])
new_x3=x3.astype(float) # int se float me change karne ke liye use karte hai
print(x3)
print(new_x3)
```
