```python
import numpy as np
```

```python
arr = np.array([1,2,3,4,5])
print("Min :-",np.min(arr)) 
print("max :-",np.max(arr))
```

```python
arr = np.array([1,2,3,4,5])
print("Min :-",np.argmin(arr))   #ye index ka min deta hai 
print("max :-",np.argmax(arr))   #ye index ka max deta hai 
```

```python
var=np.array([[1,2,3,4],[5,6,7,8]])
print(var)
print("har column ka minimum value:- \n",np.min(var,axis=0))
```

```python
var=np.array([[1,2,3,4],[5,6,7,8]])
print(var)
print("har row ka minimum value:- \n",np.min(var,axis=1))
```

```python
var=np.array([[1,2,3,4],[5,6,7,8]])
print(var)
print("har column ka max value:- \n",np.max(var,axis=0))
```

```python
var=np.array([[1,2,3,4],[5,6,7,8]])
print(var)
print("har row ka max value:- \n",np.max(var,axis=1))
```
