```python
# indexing , slicing in numpy Array
# Indexing aur Slicing NumPy me data access karne ka tareeka hai
```

```python
import numpy as np

arr = np.array([10,20,30,40])
print(arr[2])
```

```python
#2D Array Indexing
arr = np.array([[1,2,3],
                [4,5,6]])

print(arr[1][2])  # row 1, column 2
```

```python
print(arr[1,2]) 
```

```python
arr = np.array([10,20,30,40,50])

print(arr[1:4])   # Rule: [start : end]  → end include nahi hota
```

```python
# Step slicing
print(arr[0:5:2])
```

```python
# 2D Slicing
arr = np.array([[1,2,3],
                [4,5,6]])

print(arr[:,1:])
```

```python
# Negative Indexing
print(arr[-1])  # Last element
```

```python
# Full Slice
arr[:]
```
