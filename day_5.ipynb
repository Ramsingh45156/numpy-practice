#create Numpy array using numpy function

```python
import numpy as np
```

```python
# create one's array  
arr=np.ones(4)
print(arr)
print(type(arr))
print(arr.dtype)
print("your array Dimension:-",arr.ndim)
```

```python
arr=np.ones((4,2))
print(arr)
print(type(arr))
print(arr.dtype)
print("your array Dimension:-",arr.ndim)
```

```python
arr=np.ones(((3,4,4)))
arr=arr.astype(int)  # ye float se int me convert kar dega 
print(arr)
print(type(arr))
print(arr.dtype)
print("your array Dimension:-",arr.ndim)
```

```python
arr=np.ones((4,5))
arr=arr.astype(int)
print(arr)
print(arr.dtype)
print("your array Dimension:-",arr.ndim)
```

```python
arr = np.full((3,4),100)   # 3 row aur 4 column aur har element 100 hoga
print(arr)
print(type(arr))
print(arr.dtype)
print("your array Dimension:-",arr.ndim)
```
