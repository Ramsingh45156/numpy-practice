#create Numpy array using numpy function

```python
import numpy as np
```

```python
# Create Empty array

arr=np.empty((3,4))    #aise likhane par garbage values fill ho jata hai 
print(arr)
```

```python
arr=np.empty((4,3),dtype=str) # pura empty print karana ho to (dtype=str) change karna hoga
                                #aur (arr.fill(" ") karna hoag

arr.fill(" ")
print(arr)
```

##arange function

```python
arr=np.arange(4)  #0 se start karke 4 tak (4 ko include nahi karta) numbers bana deta hai.
print(arr)
```

#Diagonal me 1 print karana

```python
arr = np.eye(3)  # identity matrix banana diagonal me 1 hota hai baaki sab jagah 0
arr = arr.astype(int)
print(arr)
print(type(arr))
print("your arry is:-",arr.ndim,"D")
```

```python
arr = np.linspace(1,10,num=5)   # 1 se 10 ke beech me 5 equal parts banayega
                                #(Gap = (10 - 1) / (5 - 1) = 9 / 4 = 2.25)
print(arr)
print("your arry is:-",arr.ndim,"D")
```

```python
arr = np.linspace(1,10,num=5)  
arr = arr.astype(int)  # yaha int me convert kar 
print(arr)
print("your arry is:-",arr.ndim,"D")
```
