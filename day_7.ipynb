#create Numpy array with Random numbers

```python
import numpy as np
```

#use random.rand(n)

```python
var=np.random.rand(4) # Ye 4 random numbers generate Sab numbers 0 se 1 ke beech (0 ≤ x < 1) hote hain
print(var)
```

```python
# Ye 2 rows aur 5 columns ka array (matrix) banata hai
# Har element random float number hota hai (0 se 1 ke beech)

var=np.random.rand(2,5) 
print(var)
```

#use randn()

```python
# Ye 5 random numbers generate karta hai
#Lekin ye numbers normal distribution (Gaussian distribution) follow karte hainvar = np.random.randn(5)

# rand() → 0 se 1 ke beech numbers
# randn() → mean = 0 aur negative + positive dono values

var = np.random.randn(5)
print(var)
```

```python
# np.random.ranf(4)
# 4 random numbers generate karta hai
# range: 0 se 1 ke beech (0 ≤ x < 1)
# output floats me hota hai

# rand() → 0–1 random numbers
# ranf() → 0–1 random numbers (same kaam 😄)
# randn() → negative + positive random numbers (center 0)

var = np.random.ranf(4)
print(var)
```

```python
#Ye 5 random integers generate karta hai
#Range hoti hai: 5 se 20 ke beech (20 include nahi hota)

# 5 → start (include hota hai)
#20 → end (exclude hota hai)
#5 → total numbers chahiye

var=np.random.randint(5,20,5)
print(var)
```
